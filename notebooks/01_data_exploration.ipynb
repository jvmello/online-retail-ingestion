# 01 — Data Exploration

Initial exploration of the `online_retail.csv` dataset before ingestion into the lakehouse.

```python
import pandas as pd

file_path = "/home/jovyan/data/online_retail.csv"

df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
df.describe(include="all").T
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")
df["Revenue"] = df["Quantity"] * df["UnitPrice"]

df[["InvoiceDate", "Quantity", "UnitPrice", "Revenue"]].head()
```

```python
df["Country"].value_counts().head(10)
```

```python
df.groupby("Country")["Revenue"].sum().sort_values(ascending=False).head(10)
```
